```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_advertising=pd.read_csv('/Users/tej/Desktop/labs/ml/lab1/advertising.csv')
df1=df_advertising.drop(columns=['Sales'])
df1=df1.dropna()
X1=df1.values
```

```python
mean1=np.mean(X1,axis=0)
A1=X1-mean1
matrix1=A1.T@A1
eigenvalues1,eigenvectors1=np.linalg.eig(matrix1)
print(eigenvalues1,eigenvectors1)
```

```python
idx1 = np.argsort(-eigenvalues1)
eigenvalues1 = eigenvalues1[idx1]
eigenvectors1 = eigenvectors1[:, idx1]

num=[1, 2, 3]
error1=[]
for n in num: 
    top1 = eigenvectors1[:, :n]
    w1=A1@top1
    
    l1=w1@top1.T
    r1=l1+mean1
    
    err1=np.sum((X1-r1)**2)
    error1.append(err1)
```

```python
plt.plot(error1,marker="*")
plt.xlabel("no of eigvectors")
plt.ylabel("reconstruction error")
plt.show
```

```python
X_advertising=X1
y_advertising=df_advertising['Sales'].values.reshape(-1,1)
m = len(y_advertising)
```

```python
X_advertising = np.c_[np.ones((m, 1)), X_advertising]
```

```python
s = int(0.8 * m)
X_train_advertising, X_test_advertising = X_advertising[:s], X_advertising[s:]
y_train_advertising, y_test_advertising = y_advertising[:s], y_advertising[s:]
```

```python
def momentum_gradient_descent(X, y, learning_rate=1e-6, iterations=1000,momentum=0.9):
    m, n = X.shape
    theta = np.zeros((n, 1))
    v = np.zeros((n, 1))
    for i in range(iterations):
        error = X @ theta - y
        gradient = (1/m) * X.T @ error
        v=momentum * v + learning_rate * gradient
        theta -= v
    return theta
```

```python
def nag(X,y, learning_rate=1e-6, iterations=1000,momentum=0.9):
    m, n = X.shape
    theta = np.zeros((n, 1))
    v = np.zeros((n, 1))
    for i in range(iterations):
        theta_temp=theta-momentum*v
        error_temp = X @ theta_temp - y
        gradient = (1/m) * X.T @ error_temp
        v=momentum * v + learning_rate * gradient
        theta -= v
    return theta
```

```python
theta_advertising_mom = momentum_gradient_descent(X_train_advertising, y_train_advertising,iterations=2000)
theta_advertising_nag=nag(X_train_advertising, y_train_advertising,iterations=2000)
```

```python
y_pred_advertising_mom = X_test_advertising @ theta_advertising_mom
y_pred_advertising_nag = X_test_advertising @ theta_advertising_nag
```

```python
def met(y_test,y_pred,theta):
    SS_res = np.sum((y_test- y_pred) ** 2)
    SS_tot = np.sum((y_test - np.mean(y_test)) ** 2)
    r2 = 1 - SS_res / SS_tot
    mse = np.mean((y_test- y_pred) ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(y_test - y_pred))
    print(theta)
    print(r2)
    print(mse)
    print(rmse)
    print(mae)
```

```python
met(y_test_advertising,y_pred_advertising_mom,theta_advertising_mom)
print("\n")
met(y_test_advertising,y_pred_advertising_nag,theta_advertising_mom)
```
